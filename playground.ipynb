```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated logits from an LLM (raw, unnormalized scores)
logits = np.array([2.0, 1.8, 1.5, 0.1, -0.5])
tokens = ['apple', 'banana', 'hat', 'chair', 'fish']

def softmax(logits, temperature=1.0):
    scaled = logits / temperature
    exp = np.exp(scaled - np.max(scaled))  # for numerical stability
    return exp / np.sum(exp)

# Test with different temperatures
temperatures = [0.5, 1.0, 1.5, 2.0]
probs_by_temp = [softmax(logits, t) for t in temperatures]

# Plotting
x = np.arange(len(tokens))
plt.figure(figsize=(10, 5))
for i, t in enumerate(temperatures):
    plt.plot(x, probs_by_temp[i], marker='o', label=f'Temp={t}')
plt.xticks(x, tokens)
plt.xlabel('Token')
plt.ylabel('Probability')
plt.title('Effect of Temperature on Softmax Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
